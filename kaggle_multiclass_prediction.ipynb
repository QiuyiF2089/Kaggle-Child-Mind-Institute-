# Multicolinear(VIF)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler

# Drop the target variable (sii) and handle missing values
X = df_train_csv_filter.drop(columns=['sii']).dropna()

# Optional: Encode categorical variables if present
X = pd.get_dummies(X, drop_first=True)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data['Feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X_scaled, i) for i in range(X_scaled.shape[1])]

# Display the VIF results
print(vif_data)
```

# Dimension Reduce(PCA)

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Step 1: Prepare the data (exclude the target variable)
X = df_train_csv_filter.drop(columns=['sii']).dropna()
y = df_train_csv_filter['sii']

# Optional: Handle categorical variables
X = pd.get_dummies(X, drop_first=True)

# Step 2: Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply PCA
pca = PCA(n_components=0.95)  # Retain 95% of the variance
X_pca = pca.fit_transform(X_scaled)

# Step 4: Explained variance ratio
explained_variance = pca.explained_variance_ratio_
print(f"Explained variance by each component: {explained_variance}")
print(f"Total number of components selected: {pca.n_components_}")

# Step 5: Integrate transformed data
X_pca_df = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(X_pca.shape[1])])
X_pca_df['sii'] = y.reset_index(drop=True)

# Display the transformed dataset
print(X_pca_df.head())
```
